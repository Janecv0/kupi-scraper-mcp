```python
from kupiapi.scraper import KupiScraper
import json
```

```python
scraper = KupiScraper()
```

```python
nealko = scraper.get_discounts_by_category('nealko-napoje')
nealko_parsed = json.loads(nealko)
```

```python
import pandas as pd
from pathlib import Path

products_root = Path('produkty')

dzusy_df = pd.read_csv(products_root / 'dzusy.csv', sep=';')
koly_df = pd.read_csv(products_root / 'koly.csv', sep=';')
caje_df = pd.read_csv(products_root / 'ledove_caje.csv', sep=';')
limonady_df = pd.read_csv(products_root / 'limonady.csv', sep=';')
sirupy_df = pd.read_csv(products_root / 'sirupy.csv', sep=';')
vody_df = pd.read_csv(products_root / 'vody.csv', sep=';')

names = set(dzusy_df['name']) | set(koly_df['name']) | set(caje_df['name']) | set(limonady_df['name']) | set(sirupy_df['name']) | set(vody_df['name'])
```

```python
rows = []

for product in nealko_parsed:
    if product['name'] in names:
        for shop, price, amount, validity in zip(product['shops'], product['prices'], product['amounts'], product['validities']):
            rows.append({
                'name': product['name'],
                'shop': shop,
                'price': price,
                'amount': amount,
                'validity': validity
            })

len(rows)
```

```python
pd.DataFrame(rows).to_csv('slevy_all.csv', index=False, sep=';')
```
